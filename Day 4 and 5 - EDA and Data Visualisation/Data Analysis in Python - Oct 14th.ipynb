## Apply method in Pandas DataFRames

```python
def alone_or_not(cols):
    SibSp = cols[0]
    Parch = cols[1]
    notalone = SibSp+Parch
    if notalone > 0:
        return 0
    else: return 1
```

```python
train['Alone'] = train[['SibSp', 'Parch']].apply(alone_or_not, axis=1)
```

```python
train.head()
```

```python
train.drop('Alone', axis=1, inplace=True)
train.head()
```

```python
train['Alone'] = (train['SibSp']==0) & (train['Parch']==0)
train.head()
```

```python
train['Alone'] = train['Alone'].astype('int32')
```
